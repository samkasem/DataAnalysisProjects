# Cars Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("2. Cars Data1.csv")
```

### Replacing all null values with the mean of column

```python
df.isnull().sum()
```

```python
#change all nulls values with mean in all columns
df = df.fillna(df.mean(numeric_only=True))
```

```python
# drop rows with empty cells
df.dropna(
    axis=0,
    how='any',
    thresh=None,
    subset=None,
    inplace=True
)
```

```python
df.isnull().sum()
```

### The count of each make in the data

```python
df['Make'].value_counts()
```

### All cars were made in Asia or Europe

```python
df[df['Origin'].isin(['Asia','Europer'])]
```

```python
df[(df['Origin'] == 'Asia') | (df['Origin'] == 'Europe')]
```

### Removing all records where weight is above

```python
df[df['Weight'] > 4000]
```

```python
df[~(df['Weight'] > 4000)]
```

### Increasing all the values in MPG_City by 3

```python
df['MPG_City']
```

```python
df['MPG_City'] = df['MPG_City'].apply(lambda x: x + 3)
```
